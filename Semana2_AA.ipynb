```python
import pandas as pd

# 1. Cargar los datos de ventas desde el archivo CSV
# y mostrar una vista previa de las primeras filas

df_ventas = pd.read_csv('ventas (1).csv', encoding="ISO-8859-1")
print("Vista previa de Ventas")
print(df_ventas.head())
```

```python
# 2. Cargar los datos de clientes desde el archivo JSON
# y mostrar la cantidad total de clientes.

df_clientes = pd.read_json('clientes.json')
print(f"\nCantidad total de clientes: {len(df_clientes)}")
```

```python
# 3.  Cargar los datos de inventario desde la hoja de cálculo Excel
# y calcular el promedio de stock disponible para todos los productos.

df_inventario = pd.read_excel('inventario.xlsx')
promedio_stock = df_inventario['stock'].mean()
print(f"\nPromedio de stock disponible: {promedio_stock:.2f}")
```
