```python
# Importación de libreria
import numpy as np
# Ejercicio 1: Crear un array de 10 elementos con ceros.
ej_1 = np.zeros(10)
print("Ejercicio 1:", ej_1)
```

```python
# Ejercicio 2: Genera un array con los números del 1 al 20.
ej_2 = np.arange(1, 21)
print("Ejercicio 2:", ej_2)
```

```python
# Ejercicio 3: Crea un array con los números pares del 2 al 50.
ej_3 = np.arange(2, 51, 2)
print("Ejercicio 3:", ej_3)
```

```python
# Ejercicio 4:Matriz identidad de 3x3
ej_4 = np.eye(3)
print("Ejercicio 4 (Identidad):\n", ej_4)
```

```python
# Ejercicio 5: Crea una matriz de 4x4 con valores aleatorios entre 0 y 1.
ej_5 = np.random.rand(4, 4)
print("Ejercicio 5:\n", ej_5)
```

```python
# Ejercicio 6: Realiza la suma elemento a elemento de los arrays a=[1,2,3], b=[4,5,6].
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
ej_6 = a + b
print("Ejercicio 6:", ej_6)
```

```python
# Ejercicio 7: Multiplica dos matrices 3x3 de enteros aleatorios entre 1 y 10.
m1 = np.random.randint(1, 11, (3, 3))
m2 = np.random.randint(1, 11, (3, 3))
ej_7 = np.dot(m1, m2) # O también m1 @ m2
print("Ejercicio 7 (Multiplicación matricial):\n", ej_7)
```

```python
# Ejercicio 8: Crea un array de 15 valores aleatorios entre 1 y 100
# e imprime su máximo, mínimo y promedio.
ej_8 = np.random.randint(1, 101, 15)
print("Ejercicio 8:", ej_8)
print(f"Máximo: {ej_8.max()}, Mínimo: {ej_8.min()}, Promedio: {ej_8.mean()}")
```

```python
# Ejercicio 9: Genera una matriz 5x5 con valores del 1 al 25 y extrae su diagonal.
matriz_5x5 = np.arange(1, 26).reshape(5, 5)
diagonal = np.diag(matriz_5x5)
print("Ejercicio 9 (Diagonal):", diagonal)
```

```python
# Ejercicio 10: Dada una matriz 5x5 de enteros aleatorios entre 1 y 20,
# calcula la suma de cada fila y columna.
matriz_random = np.random.randint(1, 21, (5, 5))
suma_filas = np.sum(matriz_random, axis=1)
suma_columnas = np.sum(matriz_random, axis=0)
print("Ejercicio 10 (Matriz):\n", matriz_random)
print("Suma de filas:", suma_filas)
print("Suma de columnas:", suma_columnas)
```
